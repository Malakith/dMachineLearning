```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

data = np.load("tensorflow_python_files\simple_network\simple_network_params.npy")
#print(data)


fig = plt.figure()
fig.set_size_inches(8, 2.5)
plt.ylabel("Accuracy")
plt.xlabel("Iteration")

line_in, = plt.plot(data[:,0], data[:,1], ':', color="red", label= "Training accuracy")
line_out, = plt.plot(data[:,0], data[:,2], '--', color="blue", label= "Validation accuracy")
out_max = plt.plot(data[np.argmax( data[:,2] ), 0], np.max(data[:,2]), '^', color="blue", label="Validation accuracy max = %2.2f" % np.max(data[:,2]))
plt.legend(handles=[line_in, line_out], loc=4, numpoints=1)
plt.show()
    
# print best 
print("OoS max = %0.4f, with Is = %0.4f, for iteration = %i" %  ( np.max(data[:,2]), 
                                            data[np.argmax( data[:,2] ), 1],
                                            data[np.argmax( data[:,2] ), 0]) )
```
